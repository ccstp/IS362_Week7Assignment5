### Choose six recent popular movies. Ask at least five people that you know (friends, family, classmates, imaginary friends) to rate each of these movies that they have seen on a scale of 1 to 5. There should be at least one movie that not everyone has seen!

### Take the results (observations) and store them somewhere (like a SQL database, or a .CSV file). Load the information into a pandas dataframe. Your solution should include Python and pandas code that accomplishes the following:

### 1. Load the ratings by user information that you collected into a pandas dataframe.

```python
# Import pandas

import pandas as pan
from pandas import DataFrame, Series

# Read the information from .csv file into pandas. Set column containing reviewer name as the index.

ratings = pan.read_csv('MovieRatings.csv', index_col=0)
```

```python
# Display the movie ratings in the dataframe

ratings.head()
```

### 2. Show the average ratings for each user and each movie.

```python
# Create a copy of the movie ratings table to calculate user and movie ratings averages

ratings_avg = ratings.copy()

ratings_avg
```

```python
# Calculate the average rating for each movie

ratings_avg.loc['Movie Average'] = ratings_avg.mean(numeric_only=True).round(2)

# Calculate the average rating given by each reviewer

ratings_avg['Reviewer Average'] = ratings_avg.mean(numeric_only=True, axis=1).round(2)

ratings_avg
```

### 3. Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.

```python
# Normalize the ratings for each user (Formula from https://bi-analytics.org/forums/topic/9-standardization-vs-normalization/). 
# Transpose dataframe to normalize ratings by user instead of by movie.

ratings_norm = ((ratings - ratings.min(numeric_only=True)) / (ratings.max(numeric_only=True) - ratings.min(numeric_only=True))).T

ratings_norm
```

```python
# Create a copy of the normalized ratings table to calculate normalized user and movie ratings averages

ratings_norm_avg = ratings_norm.copy()

# Calculate the average normalized rating for each movie

ratings_norm_avg.loc['Normalized Reviewer Average'] = ratings_norm.mean(numeric_only=True).round(2)

# Calculate the average normalized rating for each user

ratings_norm_avg['Normalized Movie Average'] = ratings_norm.mean(numeric_only=True, axis=1).round(2)
ratings_norm_avg
```

## Conclusion

### 4. Provide a text-based conclusion: Explain what might be advantages and disadvantages of using normalized ratings instead of the actual ratings.

Normalization refers to the technique of scaling data so that they will fall in the range of [0,1]. 

The advantage of normalized ratings is that it makes it easier to compare the reviewers' movies ratings by ensuring that all the data are on the same scale. The normalized ratings for each reviewer show how the reviewer rates each movie relative to how others reviewers rated the same movie.

The disadvantage of normalized ratings is that outliers will result in the ratings being scaled to a very small interval. This makes it difficult to see how the movies actually compared to each other.

### 5. Create another new pandas dataframe, with standardized ratings for each user. Once again, show the average ratings for each user and each movie.

```python
# Standardize the ratings for each user (Formula from https://bi-analytics.org/forums/topic/9-standardization-vs-normalization/)
# Transpose dataframe to standardize ratings by user instead of by movie

ratings_standardized = ((ratings - ratings.mean()) / ratings.std()).T

ratings_standardized
```

```python
# Create a copy of the standardized ratings table to calculate normalized user and movie ratings averages

ratings_standardized_avg = ratings_standardized.copy()

# Calculate the average standardized rating for each movie

ratings_standardized_avg.loc['Standardized Reviewer Average'] = ratings_standardized.mean(numeric_only=True).round(2)

# Calculate the average standardized rating for each user

ratings_standardized_avg['Standardized Movie Average'] = ratings_standardized.mean(numeric_only=True, axis=1)

ratings_standardized_avg
```
